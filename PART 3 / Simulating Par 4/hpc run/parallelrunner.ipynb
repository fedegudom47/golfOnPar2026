```python
from multiprocessing import get_context
from separatesimulationrunner import run_simulation_wrapper
import pandas as pd

# --- Define water hazard Y positions ---
water_locations = list(range(100, 131, 10))  # every 10 yards from 100 to 230

# --- Parallel processing ---
ctx = get_context("spawn")  # safer in notebooks/macOS
with ctx.Pool(processes=6) as pool:  # adjust number of processes
    results = pool.map(run_simulation_wrapper, water_locations)

# --- Save results to CSV ---
results_df = pd.DataFrame(results)
results_df.to_csv("summary_of_water_tests.csv", index=False)
results_df.head()

```
